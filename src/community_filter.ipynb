```python
import matplotlib.pyplot as plt 

def number_of_communities_per_tweet_count(d):
    """pass in a dictionary of 
    number of unique tweets -> 
    total unique tweets tweeted by communities having that number of unique tweets"""
    tweet_count_2_comms = {}
    for k in d:
        num_communities = d[k]/k
        tweet_count_2_comms[k] = num_communities
    return tweet_count_2_comms

count_maps = [
    phase1_community_tweet_counts,
    phase2_community_tweet_counts,
    phase3_community_tweet_counts,
    phase4_community_tweet_counts,
    phase5_community_tweet_counts,
    phase6_community_tweet_counts,
]

count_2_comms_map = {}
for i, count_map in enumerate(count_maps):
    period = i + 1
    count_2_comms_map[period] = number_of_communities_per_tweet_count(count_map)
```

```python
def analyze_tweet_counts(threshold, count_map):
    """Returns Tuple[qty OTs included, qty OTs excluded]"""
    included, excluded = 0, 0
    for ot in count_map:
        if ot < threshold:
            excluded += count_map[ot]
        else:
            included += count_map[ot]
    return included, excluded

def analyze_num_communities(threshold, num_comms_map):
    """Returns Tuple[qty communities included, qty communities excluded]"""
    included, excluded = 0, 0
    for ot in count_map:
        if ot < threshold:
            excluded += count_map[ot]
        else:
            included += count_map[ot]
    return included, excluded    

included_OTs, excluded_OTs = 0, 0
for i in range(6):
    included, excluded = analyze_tweet_counts(thresholds[i], count_maps[i])
    included_OTs += included
    excluded_OTs += excluded

included_comms, excluded_comms = 0, 0
for i in range(6):
    included, excluded = analyze_num_communities(thresholds[i], count_2_comms_map[i + 1])
    included_comms += included
    excluded_comms += excluded

print("OTs: included = ", included_OTs, "; excluded = ", excluded_OTs, "; total = ", included_OTs + excluded_OTs)
print("Comms: included = ", included_comms, "; excluded = ", excluded_comms, "; total = ", included_comms + excluded_comms)
```
